```python
import pandas as pd
import numpy as np
```

```python
years = 2015 + np.array(range(5))
```

```python
df_main = pd.read_csv("../data/budget-%d.csv" % 2015)
cols = list(filter(lambda x: x != 'budget', df_main.columns))
df_main = df_main[cols]

for y in years:
    df = pd.read_csv("../data/budget-%d.csv" % y)[['id', 'budget']]
    df = df.rename(index=str, columns={'budget': 'budget-%d' % y})
    
    df_main = df_main.merge(df, on="id", how="left")
```

```python
df_main[
    [
    'id', 'name', 'amphur', 'province', 'type', 'size', 'area'
    ] + list(map(lambda x: 'budget-%d' % x, years))
].to_csv("../data/yearly-budget-merged.csv", header=True, index=False)
```
