```python
import json
import glob
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
files = glob.glob("../outputs/projects/*.json")
```

```python
def read_files(files):
    basket = []
    for path in files:
        with open(path, "r") as f:

            for data in json.load(f):
                winner = data['contract'][0]
                data['project_money'] = float(data['project_money'].replace(",", ""))
                data['sum_price_agree'] = float(data['sum_price_agree'].replace(",", ""))
                del data['contract']
                new_data = dict(
                    **data,
                    **winner
                )

                basket.append(new_data)

    df = pd.DataFrame(basket)
    return df
df = read_files(files)
df = df[~df.winner_tin.str.contains("-")] #todo manually map this thing
df['tin'] = df['winner_tin'].apply(lambda x: int(x.replace(".", "")))
```

```python
"we have %d records" % len(df)
```

```python
df_province_region = pd.read_csv("../data/province-region.csv")
df_comprofile = pd.read_csv("../company_profiles_and_board_list/company_profile.csv").add_prefix("p_")

df_comprofile['tin'] = df_comprofile['p_tin']
```

```python
df_merged = df.merge(df_province_region, on="province").merge(df_comprofile, how="left", on="tin")
```

```python
def extract_local_region_project(name, province):
    df_local_region = df_merged[(df_merged.dept_sub_name == name) & (df_merged.province==province)]
    print(name, province)
    print("have %d projects" % len(df_local_region))
    print("have %d orgs involved" % len(set(df_local_region.corporate_name)))
    
    data_local_region = []

    for k, v in df_local_region.groupby(["corporate_name", "purchase_method_name"])["sum_price_agree"].sum().to_dict().items():

        source = k[1]
        target = k[0] 


        data_local_region.append(dict(
            source=source, target=target, value=v
        ))

    with open("../data/viz/%s-%s.json" % (name, province), "w") as f:
        json.dump(data_local_region, f, ensure_ascii=False)
              
extract_local_region_project("เทศบาลตำบลตาพระยา", "สระแก้ว")
```

```python
def extract_data():
    data = []
    for i, c in enumerate(corp_20m):
        df_c = df_merged[df_merged.corporate_name == c]
        com = df_c.to_dict("records")[0]
        
        dd = dict(oneCategory=0, name=c)
        dd['totalProjectBudget'] = df_c.sum_price_agree.sum()
        dd['totalProjects'] = len(df_c)
        dd['regionCounts'] = sorted(df_c.groupby("region")["contract_no"].count().to_dict().items(), key=lambda d: d[1], reverse=True)
        
        methods = df_c.groupby("purchase_method_name")["contract_no"].count().to_dict()
        dd['methodCounts'] = sorted(methods.items(), key=lambda d: d[1], reverse=True)
        dd['projects'] = df_c[['sum_price_agree', 'winner_tin', 'purchase_method_name', 'province']].to_dict("records")
        dd['specificProjects'] = methods.get("คัดเลือก", 0) + methods.get("พิเศษ", 0) + methods.get("เฉพาะเจาะจง", 0)
        dd['majorityRegion'] = dd['regionCounts'][0][0]
        dd['tin'] = com['winner_tin']
        if not (type(com["p_company_nm"]) == str):
            print(com)
        
        for k in "company_nm,company_type,company_status,bus_type_cd,bus_type_nm,province,reg_capital,income,profit_loss,total_asset,equity".split(","):
            v = com["p_" + k]
            if not (type(v) == str):
                dd[k] = ""
            else:
                dd[k] = v
            
            
        data.append(dd)

    with open("/Users/heytitle/projects/jourd-politics/static/data.json", "w") as f:
        f.write(json.dumps(data, ensure_ascii=False))

extract_data()
```
